```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Revenue': [10, 15, 7, 10, 12],
    'Profit': [2, 5, 1, 3, 4]
}

df = pd.DataFrame(data)
```

```python

plt.figure(figsize=(8, 5))

plt.bar(df['Year'], df['Revenue'], label='Revenue')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Revenue by Year')
plt.legend()

plt.grid(True)

plt.show()
```

```python

plt.figure(figsize=(8, 5))

plt.plot(df['Year'], df['Profit'], marker='o', linestyle='-', color='b', label='Profit')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Profit by Year')
plt.legend()

plt.grid(True)

plt.show()
```
